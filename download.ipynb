```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
import IPython.display as display

def rgb2gray(img):
    assert (len(img.shape) == 3)  # 3D array
    return 0.3 * img[:, :, 0] + 0.59 * img[:, :, 1] + 0.11 * img[:, :, 2]

def clahe_equalized(img):
    assert (len(img.shape) == 2)  # 2D array for grayscale image
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    return clahe.apply(img)

def adjust_gamma(img, gamma=1.0):
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
    return cv2.LUT(img, table)

def adjust_brightness(img, brightness=0):
    return np.clip(img + brightness, 0, 255).astype(np.uint8)

def adjust_sharpness(img, sharpness=1.0):
    kernel = np.array([[0, -1, 0],
                       [-1, 5 + sharpness, -1],
                       [0, -1, 0]])
    return cv2.filter2D(img, -1, kernel)

def my_PreProc(image, contrast=1.0, gamma=1.0, brightness=0, sharpness=0):
    gray_image = rgb2gray(image)
    clahe_image = clahe_equalized(gray_image.astype(np.uint8))
    adjusted_image = adjust_gamma(clahe_image, gamma)
    brightness_image = adjust_brightness(adjusted_image, brightness)
    final_image = adjust_sharpness(brightness_image, sharpness)
    return final_image

def display_image(image):
    plt.imshow(image, cmap='gray')
    plt.axis('off')
    plt.show()

# Đọc ảnh PNG
image_path = '/home/khanhle/SkinSeg/SkinSeg/proceeded_png/drive/images/21.png'  # Thay đổi đường dẫn tới ảnh của bạn
image = cv2.imread(image_path)

# Hàm tương tác
def interactive_display(contrast=1.0, gamma=1.0, brightness=0, sharpness=0):
    preprocessed_image = my_PreProc(image, contrast, gamma, brightness, sharpness)
    display_image(preprocessed_image)

# Tạo thanh trượt cho gamma, độ tương phản, độ sáng, độ sắc nét
gamma_slider = FloatSlider(value=1.0, min=0.1, max=3.0, step=0.1, description='Gamma:')
contrast_slider = FloatSlider(value=1.0, min=0.0, max=3.0, step=0.1, description='Contrast:')
brightness_slider = FloatSlider(value=0, min=-100, max=100, step=1, description='Brightness:')
sharpness_slider = FloatSlider(value=0, min=0, max=5, step=0.1, description='Sharpness:')

interact(interactive_display, contrast=contrast_slider, gamma=gamma_slider, 
         brightness=brightness_slider, sharpness=sharpness_slider)
```
